```python
import cv2
```

```python
device = cv2.VideoCapture(0)
```

```python
# hd
width = 1080
height = 720
device.set(cv2.CAP_PROP_FRAME_WIDTH, width)
device.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
```

```python
flag, image = device.read()
image = cv2.flip(image, 1)
```

```python
flag
```

```python
new_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
flag
```

```python
cv2.imshow('myimg', image)
```

```python
import matplotlib.pyplot as plt
```

```python
def img_show(img):
    plt.figure(dpi=100)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
img_show(image)
```

```python
img_show(cv2.flip(new_img, 1))
```

```python
cv2.imwrite('myimg.jpg', image)
```
